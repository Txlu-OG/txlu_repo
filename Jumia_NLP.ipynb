```python
Cookie={} #insert request cookies within{}
def GetJumiaSearch(search):
    url= 'https://www.jumia.com.ng/catalog/?q='+search
    print(url)

    page=requests.get(url,headers=header)

    if page.status_code==200:
        return page

    else:
        return 'Error'
```

```python
# Function To get product details using Unique product ID number 
def SearchProductNumber(product):
    url= 'https://www.jumia.com.ng/danami-grace-fortified-hooded-short-gown-pink-'+product
    print(url)
    
    page=requests.get(url,headers=header)
    if page.status_code==200:
        return page
    else:
        return 'Error'
```

```python
def SearchProductReview(review_link):
    url='https://www.jumia.com.ng/'+review_link
    print(url)

    page=requests.get(url,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
product_names = []
response = GetJumiaSearch('gown')
if response != 'Error':
    soup = BeautifulSoup(response.content)
    for i in soup.findAll('h3',{'class':'name'}):
        product_names.append(i.text.strip())
```

```python
print(f"Product Names: {product_names}")
print(f"Number of Products: {len(product_names)}")
```

```python
product_numbers = []
response = GetJumiaSearch('gown')
if response != 'Error':
    soup = BeautifulSoup(response.content, 'html.parser')
    for div in soup.findAll('div', {'class': 'img-c'}):
        img_tag = div.find('img')
        if img_tag:
            data_src = img_tag.get('data-src')
            if data_src:
                # Extract the unique product identifier from the URL
                try:
                    product_number = data_src.split('/product/')[1].split('/')[1]  # Corrected extraction logic
                    product_numbers.append(product_number)
                except IndexError:
                    print(f"Error processing URL: {data_src}")
```

```python
print(f"Product Numbers: {product_numbers}")
print(f"Number of Product Numbers: {len(product_numbers)}")
```

```python
links = []
for i in range(len(product_numbers)):
    response = SearchProductNumber(product_numbers[i])
    if response != 'Error':
        soup = BeautifulSoup(response.content, 'html.parser')
        for a_tag in soup.findAll('a', {'class': '-plxs _more'}):
            links.append(a_tag['href'])
```

```python
print(f"Review Links: {links}")
```

```python
all_reviews = []
```

```python
def ParseReviews(soup):
    reviews = []
    review_section = soup.find('div', {'class': 'card mrg-btm'})
    if review_section:
        for review in review_section.findAll('article', {'class': 'card'}):
            review_text = review.find('p', {'class': 'rev-text'})
            review_rating = review.find('div', {'class': 'stars'})
            if review_text and review_rating:
                review_content = review_text.text.strip()
                review_rating = review_rating['style'].split(':')[1].strip('%')
                reviews.append({'review_content': review_content, 'review_rating': int(review_rating) // 20})
    return reviews

for link in links:
    review_url = 'https://www.jumia.com.ng' + link
    response = SearchProductReview(review_url)
    if response != 'Error':
        soup = BeautifulSoup(response.content, 'html.parser')
        product_reviews = ParseReviews(soup)
        all_reviews.extend(product_reviews)
```

```python
print(f"Collected Reviews: {all_reviews}")

df_reviews = pd.DataFrame(all_reviews)
print(df_reviews)
```
