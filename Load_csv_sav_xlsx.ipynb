```python
import pandas as pd

# Function to load and view the file based on the file extension
def load_and_view_file(file_path):
    # Check the file extension and load the file accordingly
    if file_path.endswith('.csv'):
        # Load CSV file
        df = pd.read_csv(file_path)
    elif file_path.endswith('.xlsx'):
        # Load Excel file
        df = pd.read_excel(file_path)
    elif file_path.endswith('.sav'):
        # Load SPSS file
        df = pd.read_spss(file_path)
    else:
        raise ValueError("Unsupported file format. Please provide a .csv, .xlsx, or .sav file.")
    
    # Display the first few rows of the dataset
    print(f"Loaded file: {file_path}")
    print(f"Number of rows and columns: {df.shape}")
    print(f"First 5 rows:\n{df.head()}")

# Provide the path to your file
file_path = '/Users/m1/Downloads/data/lassa/Lassa Fever_Dataset_NCDC.sav' 

# Load and view the file
load_and_view_file(file_path)
```
